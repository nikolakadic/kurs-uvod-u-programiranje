```python
# args i kwargs

def foo(*b):
    product = 1
    for el in b:
        product *= el
    return product

def boo(**kwargs):
    print(kwargs)
    for keyword, value in kwargs.items():
        print(keyword, '->', value)
    
```

```python
foo(3,6,5)
```

```python
boo(ime='Janko',prezime='Jankovic',godine=25)
```
